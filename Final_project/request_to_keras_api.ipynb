```python
import os
import requests
import numpy as np
```

```python
import matplotlib.pyplot as plt
from keras.preprocessing.image import load_img
```

```python
KERAS_API_URL = "http://127.0.0.1:5000/predict"
```

```python
images_path = {img.split(".")[0] :os.path.join(os.getcwd(), "images_for_predictions", img) 
               for img in os.listdir(os.path.join(os.getcwd(), "images_for_predictions"))}
```

```python
for k,v in images_path.items():
    image = open(v, "rb").read()
    r = requests.post(KERAS_API_URL, files={"image": image}).json()
    if r["success"]:
        print("="*100)
        predicted_label = r["predicted_label"]
        print(f"Model predict {predicted_label} label", 
              f"True label is {k}", sep="\n")
        img = load_img(v)
        img = np.array(img)
        plt.imshow(img)
        plt.show()
        
```
